```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from apyori import apriori
```

```python
dataset = pd.read_csv('Order2.csv', header = None)
dataset
```

```python
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
# Training Apriori algorithm on the dataset
rule_list = apriori(transactions, min_support = 0.003, min_confidence = 0.5, min_lift = 3, min_length = 2)
```

```python
# Visualizing the list of rules
results = list(rule_list)
for i in results:
    print('\n')
    print(i)
    print('**********') 
```
